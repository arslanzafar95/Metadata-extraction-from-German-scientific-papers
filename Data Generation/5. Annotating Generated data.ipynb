```python
########### dataframe cleaning and preprocessing ###########################

import pandas as pd
df = pd.read_excel("dataframe.xlsx")
df

counter=0
for x in df['author']:
    df['author'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").strip()
    #print(df['author'][counter])
    counter = counter+1
#df['author'][0].replace("[","").replace("'","").replace("]","")

counter=0
for x in df['year']:
    df['year'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").strip()
    #print(df['year'][counter])
    counter = counter+1
   
counter=0
for x in df['title']:
    df['title'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").replace('"',"").strip()
    #print(df['title'][counter])
    counter = counter+1
    
counter=0
for x in df['doi']:
    df['doi'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").replace('"',"").strip()
    #print(df['doi'][counter])
    counter = counter+1
      
counter=0
for x in df['journal']:
    df['journal'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").replace('"',"").strip()
    #print(df['journal'][counter])
    counter = counter+1
    
counter=0
for x in df['publisher']:
    df['publisher'][counter]=x.replace("[","").replace("'","").replace("]","").replace("0001","").replace("0002","").replace('"',"").strip()
    #print(df['publisher'][counter])
    counter = counter+1
    
df=df.replace('',"na")
df.head()
```

```python
# run this file in the same folder where images and pdfs are placed
# change the file names to the templates you are working on, like here template is 10939, so just replace 10939


number=0
while True:

    
    names='10939_'+str(number)
    
    # change the number limit to number of files you are generating
    if number==564:
        break

    import pdfquery
    # loading a pdf
    pdf = pdfquery.PDFQuery(names+".pdf")
    pdf.load()
    # if you want, you can store the entire pdf as an xml file

    pdf.tree.write(names+".xml", pretty_print=True, encoding="utf-8")

    import xml.etree.ElementTree as ET
    tree = ET.parse(names+'.xml')
    root = tree.getroot()

   
    author=df['author'][number]
    title=df['title'][number]
    journal=df['journal'][number]
    #doi=df['doi'][number]
    #year=df['year'][number]
    
    # create lists for all types of metadata bounding boxes 
    bounding_box_list=[]
    author_list=[]
    title_list=[]
    journal_list=[]
    #doi_list=[]
    #year_list[]

    # parse xml 
    for child in root:
        print('START')
        print (child.tag,"----", child.attrib)
        
        
        for fchild in child:
            print (fchild.tag,"----", fchild.attrib)
            for ffchild in fchild:
                print (ffchild.tag,"----", ffchild.attrib)
                print('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
                print(ffchild.text)
                print(ffchild.attrib['x0'])

                bounding_box_list.append([ffchild.text.strip(),[ffchild.attrib['x0'],ffchild.attrib['x1'],ffchild.attrib['y0'],
                                                                ffchild.attrib['y1']]])

                if ffchild.text.strip() in author and ffchild.text.strip() is not '':
                    author_list.append([ffchild.text.strip(),[ffchild.attrib['x0'],ffchild.attrib['x1'],ffchild.attrib['y0'],
                                                              ffchild.attrib['y1']]])
                elif ffchild.text.strip() in title and ffchild.text.strip() is not '':
                    title_list.append([ffchild.text.strip(),[ffchild.attrib['x0'],ffchild.attrib['x1'],ffchild.attrib['y0'],
                                                             ffchild.attrib['y1']]])
                elif ffchild.text.strip() in journal and ffchild.text.strip() is not '':
                    journal_list.append([ffchild.text.strip(),[ffchild.attrib['x0'],ffchild.attrib['x1'],ffchild.attrib['y0'],
                                                               ffchild.attrib['y1']]])
                #elif ffchild.text.strip() in doi and ffchild.text.strip() is not '':
                #    doi_list.append([ffchild.text.strip(),[ffchild.attrib['x0'],ffchild.attrib['x1'],ffchild.attrib['y0'],
                #                                              ffchild.attrib['y1']]])

                print('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
            print('END')
            print('\n')

    coordinates_author = []
    coordinates_journal = []
    coordinates_title = []
    #coordinates_doi = []

    for lst in author_list:
        bboxes=lst[1]
        print((bboxes))
        print('\n')
        apt1=[float(bboxes[0]),842-float(bboxes[3])]
        apt2=[float(bboxes[0]),842-float(bboxes[2])]
        apt3=[float(bboxes[1]),842-float(bboxes[2])]
        apt4=[float(bboxes[1]),842-float(bboxes[3])]
        coordinates_author.append([ apt1,apt2,apt3,apt4 ])
    print('-----------')
    print(coordinates_author)

    for lst in title_list:
        bboxes=lst[1]
        print((bboxes))
        print('\n')
        tpt1=[float(bboxes[0]),842-float(bboxes[3])]
        tpt2=[float(bboxes[0]),842-float(bboxes[2])]
        tpt3=[float(bboxes[1]),842-float(bboxes[2])]
        tpt4=[float(bboxes[1]),842-float(bboxes[3])]
        coordinates_title.append([ tpt1,tpt2,tpt3,tpt4 ])
    print('-----------')
    print(coordinates_title)

    for lst in journal_list:
        bboxes=lst[1]
        print((bboxes))
        print('\n')
        jpt1=[float(bboxes[0]),842-float(bboxes[3])]
        jpt2=[float(bboxes[0]),842-float(bboxes[2])]
        jpt3=[float(bboxes[1]),842-float(bboxes[2])]
        jpt4=[float(bboxes[1]),842-float(bboxes[3])]
        coordinates_journal.append([ jpt1,jpt2,jpt3,jpt4 ])
    print('-----------')
    print(coordinates_journal)
    
    #for lst in doi_list:
    #    bboxes=lst[1]
    #    print((bboxes))
    #    print('\n')
    #    dpt1=[float(bboxes[0]),842-float(bboxes[3])]
    #    dpt2=[float(bboxes[0]),842-float(bboxes[2])]
    #    dpt3=[float(bboxes[1]),842-float(bboxes[2])]
    #    dpt4=[float(bboxes[1]),842-float(bboxes[3])]
    #    coordinates_doi.append([ dpt1,dpt2,dpt3,dpt4 ])
    #print('-----------')
    #print(coordinates_doi)


    lst=[coordinates_author,coordinates_title,coordinates_journal]
    lst
    
   
    # create a list of all metadata bounding boxes and use in JSON as 'shapes' value
    # abase --> for authors, tbase --> for titles and so on
    abase = {
          "label": "author",
          "points": [       [250,269],[250,283],[329,283],[329,269]         ],
          "shape_type": "polygon", "flags": {}
        }
    a1base={
          "label": "author",
          "points": [       [250,269],[250,283],[329,283],[329,269]         ],
          "shape_type": "polygon", "flags": {}
        }
    a2base={
          "label": "author",
          "points": [       [250,269],[250,283],[329,283],[329,269]         ],
          "shape_type": "polygon", "flags": {}
        }
    tbase = {
          "label": "title",
          "points": [        [197,332],[197,350],[383,350],[383,332]      ],
           "shape_type": "polygon","flags": {}
        }
    t1base={
          "label": "title",
          "points": [        [197,332],[197,350],[383,350],[383,332]      ],
           "shape_type": "polygon","flags": {}
        }
    t2base={
          "label": "title",
          "points": [        [197,332],[197,350],[383,350],[383,332]      ],
           "shape_type": "polygon","flags": {}
        }
    jbase = {
          "label": "journal",
          "points": [        [202,381],[202, 395],[378, 395],[378, 381]      ],
          "shape_type": "polygon","flags": {}
        }
    j1base = {
          "label": "journal",
          "points": [        [202,381],[202, 395],[378, 395],[378, 381]      ],
          "shape_type": "polygon","flags": {}
        }
    
    ab=[abase,a1base,a2base]
    tb=[tbase,t1base,t2base]
    jb=[jbase,j1base]

    
    liste=[]
    try:
        counter=0
        for no in range(len(coordinates_author)):   

            ab[counter]['points']=lst[0][counter]
            print(ab[counter])
            liste.append(ab[counter])
            counter=counter+1

        counter=0
        for no in range(len(coordinates_title)):   

            tb[counter]['points']=lst[1][counter]
            print(tb[counter])
            liste.append(tb[counter])
            counter=counter+1    

        counter=0
        for no in range(len(coordinates_journal)):   

            jb[counter]['points']=lst[2][counter]
            print(jb[counter])
            liste.append(jb[counter])
            counter=counter+1
    except:
        print('============================================================================')
        print(number)
        number=number+1
        continue

    import base64
    import json
    # get image data 
    file=names+".jpeg"
    image_data = base64.b64encode(open(file, "rb").read())
    imagedata=str(image_data)[1:].replace("'","")

    x ='''[{
      "version": "4.4.0",  "flags": {},

      "shapes": [
        {
          "label": "author",
          "points": [       [250,269],[250,283],[329,283],[329,269]         ],
          "group_id": null,"shape_type": "polygon", "flags": {}
        },

        {
          "label": "title",
          "points": [        [197,332],[197,350],[383,350],[383,332]      ],
          "group_id": null, "shape_type": "polygon","flags": {}
        },

        {
          "label": "journal",
          "points": [        [202,381],[202, 395],[378, 395],[378, 381]      ],
          "group_id": null,"shape_type": "polygon","flags": {}
        }  ],

      "imagePath": "10939.jpg",
      "imageData": "/9j",
      "imageHeight": 842,
      "imageWidth": 595
    }]'''

    try:
        info=json.loads(x)
        print('User count', len(info))
        print("-------------")
        
        # replace shapes value in x JSON 
        info[0]['shapes']=liste


        print(info[0]['shapes'])
        info[0]['imageData']=imagedata
        print("-------------------------------------------------------------------------------------------")




        #info[0]['shapes'][0]['points']=lst[0][0]
        #info[0]['shapes'][1]['points']=lst[1][0]
        #info[0]['shapes'][2]['points']=lst[2][0]
        #print(key['shapes'])
        #info[0]['imageData']=imagedata
        #print("-------------------------------------------------------------------------------------------")



        for key in info:
            my_details=key
        name=names+".json"
        with open(name, 'w') as json_file:
            json.dump(my_details, json_file)
    except:
        
        print('============================================================================')
        print(number)
        number=number+1
        continue
    
        

    print('============================================================================')
    print(number)
    number=number+1
```

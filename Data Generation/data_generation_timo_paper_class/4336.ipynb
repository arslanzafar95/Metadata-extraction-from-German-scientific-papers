# Annotate all pdfs of template 4336

```python
from glob import glob
```

```python
pdfs = glob("4336_PDFs/*.pdf")
pdfs
```

```python
errors = []
for pdf in pdfs:
    
    print(pdf)
    
    p = Paper(pdf, {"title": "", "author":"", "abstract": "", "date": "6/", "journal": ""})
    try:
        p.resize_image(569,794)
        p.save_image("4336_annotations/")
        p.get_metadata_items(metadata_df)
        p.bbox_from_text()
        #print(p.rectangles)
        p.get_annotations()
        p.save_annotations("4336_annotations/")
    except:
        errors.append(pdf)
        pass
```

```python
errors
```

```python
import os
for pdf in errors:
    print(pdf[10:-4] + ".jpeg")
    jpeg = "4336_annotations/" + pdf[10:-4] + ".jpeg"
    json = "4336_annotations/" + pdf[10:-4] + ".json"
    if os.path.exists(jpeg):
        os.remove(jpeg)
        print("removed jpeg: " + jpeg)
    if os.path.exists(json):
        os.remove(json)
        print("removed json: " + json)
    
```
